## Sequence Investigation
Sequences for each of the specific r values.

---

### $r=2.5$

```python
a1 = mid.SequencePlotter(r=2.5)
a1.plot()
a2 = mid.SequencePlotter(r=2.5, x0=0.1)
a2.plot()
```

```python
a1.xs[-10:], a2.xs[-10:]
```

#### Behavior
The first graph starting at $x=0.5$ shows a small deviation at the start and then levels out at a population value of $0.6$. The deviations once it levels out are at $10^{-15}$ which are negligable.
The second graph starting at $x=0.1$ shows the same trend. Although the starting point was lower than the first graph, this does not seem to have an overall effect on the end population.

### $r=3.2$

### $r=3.5$

```python
c1 = mid.SequencePlotter(r=3.5)
c1.plot()
c2 = mid.SequencePlotter(r=3.5,x0=0.1)
c2.plot()
```

```python
c1.xs[-10:], c2.xs[-10:]
```

#### Behavior
The deviations at this r value are unique to what we have seen so far. The pattern of deviation, instead of alternating between two populations, repeats itself every four population values: $0.82694, 0.50088, 0.87499, 0.38281$. This pattern makes the output look like two different graphs; one smaller deviating graph inside of a larger deviating graph. Again we see that both graphs show the same alternations despite different initial x conditions.

### $r=3.5441$

```python
d1 = mid.SequencePlotter(r=3.5441)
d1.plot()
d2 = mid.SequencePlotter(r=3.5441,x0=0.1)
d2.plot()
```

```python
d1.xs[-10:], d2.xs[-10:]
```

#### Behavior
This graph, although initially very similar to $r=3.5$, has twice as many alternations per repeition: $0.81724, 0.52932, 0.88297, 0.36620, 0.82258, 0.51722, 0.88497, 0.36077$. Although the r value was increased by only $0.441$, the alternations per repetition of values doubled! It is starting to become apparent that as the r value increases alternations between the same values becomes less frequent. Again, despite different starting x conditions, the graphs alternate between the same values although at different iterations of k.

### $r=3.5699$

```python
e1 = mid.SequencePlotter(r=3.5699)
e1.plot()
e2 = mid.SequencePlotter(r=3.5699,x0=0.1)
e2.plot()
```

```python
e1.xs[-10:], e2.xs[-10:], e1.xs[-20:]
```

#### Behavior
These pair of graphs do not show any pattern within the last 10 points. It now takes 16 alternations until the pattern is repeated. This is showing an even larger pattern now: as the r values increase, the population will alternate at double of the previous alternation (1,2,4,8,16). With greater r values the population graph becomes more chaotic and random.

---

### $r=3.57$

```python
f1 = mid.SequencePlotter(r=3.57)
f1.plot()
f2 = mid.SequencePlotter(r=3.57,x0=0.1)
f2.plot()
```

```python
f1.xs[-10:], f2.xs[-10:], f1.xs[-35:]
```

#### Behavior
As predicted from the behavior at the last pair of graphs at $r=3.5699$, the number of alternations per repetition of values has doubled, now at 32! Previously, the r values needed to increase much more for the pattern to double. Starting at $r=2.5$, the value increased by $1$. Now, the value of r only needs to increase far less in order for the this pattern to double; $r=5.5699$ to $5.7$. If this trend continues, the alternations of the population value will become so far apart that it will seem random.

## **Scatterplot**

```python
mid.scatterplot()
```

#### Behavior and Conlusion
The scatterplot shows each asymptotic population along the range of r values $[2.9, 4]$ with iterations of $0.001$. This gives us a very clear representation of the populations behavior at every possible value of r to the thousandth along this range.


This graph shows exactly what was predicted in the behaviors of the previous sequence graphs. At early values of r, there is relatively no alternations, hence the singular horizontal line. Once the r values reach around $r=3.0$, they begin to alternate between two values, shown by the branch in the horizontal line. As the r values increase, each branch repeats this behavior and at smaller intervales of r. Once the r values are large enough each branch has repeated this process more times and much more frequently. The graph begins to look very chaotic and random. As far as the eye can see their is no clear pattern to the points after around $r=3.6$.


This behavior would have been hard to predict from just the equation, but becomes more clear once you can visually see the patterns at different r values.
